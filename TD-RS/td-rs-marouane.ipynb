```python
# TP : Réseaux sémantiques
# Nom : Marouane Boufarouj
# Date : 11-11-2024

"""
Objectif : Création et visualiser un réseau sémantique simple
représentant les relations entre différents concepts.
"""
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
concepts = ['chien', 'animal', 'viande', 'nourriture', 
           'chat', 'être vivant', 'respirer']
G.add_nodes_from(concepts)
relations = [
    ('chien', 'animal', 'est un'),
    ('chien', 'viande', 'mange'),
    ('viande', 'nourriture', 'est une'),
    ('chat', 'animal', 'est un'),
    ('animal', 'être vivant', 'est un'),
    ('animal', 'respirer', 'a besoin de')
]

G.add_edges_from((x, y) for x, y, z in relations)
etiquettes_relations = {(x, y): z for x, y, z in relations}
plt.figure(figsize=(12, 8))

disposition = nx.spring_layout(G, k=1, iterations=50)
nx.draw(G,disposition,node_color='lightblue',node_size=2000,with_labels=True,font_size=10,font_weight='bold',font_color='black',arrows=True)
nx.draw_networkx_edge_labels(G, disposition, edge_labels=etiquettes_relations)

plt.title("Réseau Sémantique - Relations entre Concepts")

plt.show()

print("\nRecherche des relations pour 'chien':")

for successeur in G.successors('chien'):
    print(f"chien {etiquettes_relations[('chien', successeur)]} {successeur}")

for predecesseur in G.predecessors('chien'):
    print(f"{predecesseur} {etiquettes_relations[(predecesseur, 'chien')]} chien")

print("\nRelation entre 'chien' et 'animal':")

if G.has_edge('chien', 'animal'):
    print(f"chien {etiquettes_relations[('chien', 'animal')]} animal")
```
